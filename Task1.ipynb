```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression  
```

```python
student_data = pd.read_csv('student_data.csv')
student_data.head()
```

```python
student_data.plot(x='Hours', y='Scores', linestyle='dashed',color='Blue')  
plt.title('Student Hours vs student Percentage')
plt.xlabel('Student take hours')
plt.ylabel('Scores')
plt.show()
```

```python
X = student_data.iloc[:, :-1].values  
y = student_data.iloc[:, 1].values  
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0) 
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
line = regressor.coef_*X+regressor.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

```python
y_pred = regressor.predict(X_test)
```

```python
pred = regressor.predict([[9.25]])

print("Hour = {}".format(9.25))
print("Score = {}".format(pred[0]))
```
